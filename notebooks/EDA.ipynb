# Exploratory Data Analysis: Flipkart Grid Traffic Demand
This notebook visualizes the spatial and temporal distribution of traffic demand across various Geohashes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
train = pd.read_csv('../dataset/train.csv')

# Parse timestamps
time_parts = train['timestamp'].str.split(':', expand=True).astype(int)
train['hour'] = time_parts[0]
train['minute'] = time_parts[1]
```

## Temporal Analysis
Visualizing how demand fluctuates throughout the day.

```python
plt.figure(figsize=(12, 5))
hourly_demand = train.groupby('hour')['demand'].mean()
sns.lineplot(x=hourly_demand.index, y=hourly_demand.values, marker='o')
plt.title('Average Traffic Demand by Hour of Day')
plt.xlabel('Hour (0-23)')
plt.ylabel('Average Demand')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## Spatial Analysis
Visualizing demand distribution across top Geohashes.

```python
plt.figure(figsize=(14, 6))
top_geohashes = train.groupby('geohash')['demand'].mean().sort_values(ascending=False).head(20)
sns.barplot(x=top_geohashes.index, y=top_geohashes.values, palette='viridis')
plt.title('Top 20 Geohashes by Average Demand')
plt.xticks(rotation=45)
plt.ylabel('Average Demand')
plt.show()
```
